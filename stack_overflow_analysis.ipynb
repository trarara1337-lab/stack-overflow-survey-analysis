Завдання 2. Аналіз повноти відповідей респондентів

```python
questions = set(schema["qname"])
columns = set(df.columns)
valid_questions = list(questions.intersection(columns))
respondents_full = df[valid_questions].notna().all(axis=1).sum()
respondents_full
```

Завдання 3. Статистичний аналіз досвіду респондентів

```python
mean_exp = df["WorkExp"].mean()
median_exp = df["WorkExp"].median()
mode_exp = df["WorkExp"].mode()[0]
pd.Series({
    "Mean":mean_exp,
    "Median":median_exp,
    "Mode":mode_exp
})
```

Завдання 5. Визначення популярності Python

```python
python_users = df["LanguageHaveWorkedWith"].str.contains("Python", na=False)

python_percent = python_users.mean() * 100

print(f"{python_percent:.2f}%")
```

Завдання 6. Аналіз шляхів навчання програмуванню

Завдання 7. Географічний аналіз компенсації Python-розробників

```python
python_dev = df[df["LanguageHaveWorkedWith"].str.contains("Python", na=False)]
python_dev = python_dev.dropna(subset=["ConvertedCompYearly"])
salary_stats = (
    python_dev
    .groupby("Country")["ConvertedCompYearly"]
    .agg(
        mean_salary="mean",
        median_salary="median"
    )
    .round(2)
)
salary_stats
```

*Завдання 9. Аналіз популярності Python по віковим категоріям 

```python
df["PythonUser"] = df["LanguageHaveWorkedWith"].str.contains("Python", na=False)
age_python = (
    df.groupby("Age")["PythonUser"]
    .mean()
    .mul(100)
    .round(2)
    .reset_index()
)
age_python
```

*Завдання 10. Аналіз індустрій серед високооплачуваних віддалених працівників

```python
salary_q75 = df["ConvertedCompYearly"].quantile(0.75)
top_remote = df[
    (df["ConvertedCompYearly"] > salary_q75) &
    (df["RemoteWork"] == "Remote")
]
industry_counts = (
    top_remote["Industry"]
    .value_counts()
)
industry_counts
```
